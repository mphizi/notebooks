# Sudoku agent

The task of this project is the same as in the `solving-agent`, 
but now the visualization uses `clingraph`:
* https://github.com/potassco/clingraph

Clingraph can be installed using conda:
```
conda install -c potassco clingraph
```

You can start from the script `sudoku-agent.py` and follow the instructions 
in the comments.
